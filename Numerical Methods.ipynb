```octave
function [L,U] = getLU(A)
    % Number of Rows or Columns of Matrix A
    n = size(A)(1);
 
    % L and U are Lower and Upper Triangle Matrix respectively.
    U = zeros(n,n);
    L = zeros(n,n);
 
    % LU decomposition using Gauss Elimination
    for i = 1:n
        for j = i:n
            if i==j
                sum = 0;
                for k = 1:j-1
                    sum += L(j,k)^2;
                end
                L(i,i) = sqrt(A(i,i) - sum);
                printf("L(%d, %d) = sqrt(%d - %d) = %d\n",i,i,A(i,i), sum,L(i,i))
                
            else
                sum = 0;
                for k = 1:j-1
                    sum += L(j,k)*L(i,k);
                end
                L(j,i) = (A(j,i) - sum)/(L(i,i));
                printf("L(%d, %d) = (%d - %d)/(%d) = %d\n",j,i,A(j,i), sum,L(i,i), L(j,i))
            end
        end
       printf("\n------------\n");
    end
    U = L';
end
[L,U] = getLU([4,2,14;2,17,-5;14,-5,83]);

function x = solve(A,M)
    [L,U] = getLU(A);
    n = size(A)(1);
    % Backward Substitution L * d = M  where d = UX
    d = zeros(n,1);
    for i = 1:n
        sigma = M(i,1);
        for j = 1:i-1
            sigma -= L(i,j)*d(j,1) ;
        end
        d(i,1) = sigma/L(i,i) ;
    end
    
    L, d
    % Backward Substitution d = U * X
    x = zeros(n,1);
    for i = n:-1:1
        sigma = d(i,1);
        for j = n:-1:i+1
            sigma -= U(i,j)*x(j,1) ;
        end
    x(i,1) = sigma/U(i,i) ;
    end
    U
end
A = [5 1 0; 1 5 1; 0 1 5];
b = [7 ;14; 17];
solve(A,b)
```

### Partial Pivoting

```octave
A = [1 2 3; 2 2 2; 3 3 3]
b = [1;2;3]

function [L,b] = doPivot(A,b)
    n = size(A)(1);
    for col = 1:n
        % Find max|a(i,col)|
        mxRow = col;
        mxVal = A(col,col);
        for i=col:n 
            if(mxVal<A(i,col))
                mxVal = A(i,col);
                mxRow = i;
            end
        end
        
        temp=A(col,:);
        A(col,:)=A(mxRow,:);
        A(mxRow,:)=temp
        tmp = b(col);
        b(col) = b(mxRow);
        b(mxRow) = tmp;
        nor = A(col,col);
        if nor!=0
            for j = col:n
                A(col,j)/=nor;
            end
            b(col)/=nor;
        end
        for i=col+1:n
            fact = A(i,col);
            for j = 1:n
                A(i,j) -= fact*A(col,j);
            end
            b(i)-=fact*b(col);
        end
    end
    L = A;
end

[L,b] = doPivot(A,b)
```

```octave
a = [1 2 1; 1 3 1; 1 1 2]

function [q,r] = QRdecompos(a)
    n = size(a)(2);
    q = zeros(n,n);
    r = zeros(n,n);
    q(:,1) = a(:,1)/norm(a(:,1),2);
    for k = 1:n-1
        sigma = 0;
        for i = 1:k
            sigma += ((a(:,k+1)')*q(:,i))*q(:,i) ;
        end
        z = a(:,k+1) - sigma;
        q(:,k+1) = z/norm(z,2);
    end
    for i = 1:n
        for j = i:n
            r(i,j) = (q(:,i)')*a(:,j);
        end
    end
end
Q = []; R = [];
for steps = 1:10
    [Q,R] = QRdecompos(a);
    a = R*Q;
end
Q , R, (Q*(Q')), Q*R
```
